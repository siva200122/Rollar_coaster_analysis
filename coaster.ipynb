KDE(Kernel Density Estimation) plot is like looking at the smooth curve made by adding up lots of tiny bumps . It helps you understand how the data is spread out without counting each individual data point.

 If the curve is tall and skinny, it means most of the scores are close together. If it's shorter and wider, it means the scores are more spread out.


y axis:  The y-axis of a KDE plot represents the density or probability density. It shows you the concentration of data points for different values on the x-axis.
 it tells you how many data points are concentrated around a specific value.

:The Gaussian kernel function looks like a bell-shaped curve, with most of its values concentrated around its mean. This makes it a good choice for smoothing data and estimating the probability density function (PDF) of a random variable in KDE.

```python
ax = df['Speed_mph'].plot(kind = 'kde', title = 'speed mph ')  #This means you're telling pandas to divide your data into 20 equally spaced bins. So, if you have a range of numbers, pandas will split that range into 20 intervals, and then it will count how many numbers fall into each interval.
ax.set_xlabel('speed mph')
```

Matplotlib:

Think of Matplotlib as your basic toolbox. It gives you tools to draw all sorts of plots, like line charts, scatter plots, and histograms.
It's like having a bunch of building blocks. You can build anything you want, but you have to assemble everything yourself, which might take a bit more effort.
You have to tell Matplotlib exactly what you want, like where to put each point and what color to make it.

Seaborn:

Seaborn, on the other hand, is like a pre-built Lego set. It's built on top of Matplotlib, so it still has all the same tools, but it comes with some pre-designed sets that you can easily use.
It's like having ready-made templates for common plots. You can quickly create things like fancy scatter plots or colorful histograms with just a few lines of code.
Seaborn also makes your plots look nicer by default, with better colors and styles, so you don't have to tweak them as much.

```python
df.plot(kind='scatter',x='Speed_mph',y='Height_ft',
        title = 'Coaster speed vs height')
plt.show()
```

 "hue" typically refers to a variable that is used to color the points based on different categories or levels of that variable
